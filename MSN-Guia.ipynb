### Problema 1

La función de Brillouin puede aproximarse en los límites de campos bajos y temperaturas altas por un comportamiento lineal. A partir de esta aproximación deduzca el valor de la constante de Curie $C = \frac{n\mu_0\mu_{\text{eff}}^2}{3k_B}$ en la expresión de la susceptibilidad magnética $ \chi = \frac{C}{T} $.

Ayudas (en el sistema SI):

$
\frac{M}{M_S} = B_J(y) = \frac{2J+1}{2J} \coth\left( \frac{2J+1}{2J} y \right) - \frac{1}{2J} \coth\left( \frac{y}{2J} \right)
$

$
y = \frac{g_J \mu_B J B}{k_B T}; \quad M_S = n g_J \mu_B J; \quad \mu_{\text{eff}} = g_J \mu_B \sqrt{J(J+1)}
$


#### Respuesta

Haciendo uso de que el desarrollo en serie de potencias para la cotangente hiperbólica es:

$coth(z) = \frac{1}{z} + \frac{z}{3}- \frac{z^3}{45}+...$

Reemplazando con los primeros dos términos en la función de Brilluin obtenemos

$$
    \begin{split}
        B_J(y) & = \frac{2J+1}{2J} \left[ \left(\frac{2J}{(2J+1)y} \right) + \left(\frac{(2J+1)y}{6J}\right) \right] - \frac{1}{2J}\left(\frac{2J}{y}+\frac{y}{6J}\right)\\
        & = \frac{(2J+1)2J}{2J(2J+1)y} + \frac{(2J+1)^2 y}{12J^2} - \frac{1}{y}-\frac{y}{12J^2}\\
        & = \cancel{\frac{1}{y}} + \frac{(2J+1)^2 y}{12J^2} -\cancel{ \frac{1}{y}}-\frac{y}{12J^2} \\
        & = \frac{y ((2J+1)^2 -1)}{12J^2}\\
        & = \frac{y (4J^2 + 4J + \cancel{1} - \cancel{1})}{12J^2}\\
        & = \frac{y4J(J+1)}{12J^2}\\
        & = \frac{y (J+1)}{3J}
    \end{split}
$$

Como tenemos campos bajos y temperaturas altas, $y \ll 1$ y $M \propto B$. Usamos:

$
    \boxed{
    M = M_S \cdot B_J(y)
    }
$

$$
    \begin{split}
        M & = n g_J\mu_B \cancel{J} \frac{g_J \mu_B J B}{k_B T} \frac{(J+1)}{3\cancel{J}}\\
        & = \frac{n g_J^2 \mu_B^2 J (J+1)}{3k_B T} B
    \end{split}
$$

Luego tenemos que:

$
    \boxed{M = \chi B}
$

$
    \boxed{\chi = \frac{C}{T}}
$

$
    \longrightarrow \chi = \frac{n g_J^2 \mu_B^2 J (J+1)}{3k_B T} = \frac{C}{T}
$

$
    \longrightarrow C = \frac{n g_J^2 \mu_B^2 J (J+1)}{3k_B}
$

Y como $\mu_{\text{eff}} = g_J \mu_B \sqrt{J(J+1)}$

$
    \mu_{\text{eff}}^2 = g_J^2 \mu_B^2 J(J+1)
$

$
    \longrightarrow \boxed{C = \frac{n \mu_0 \mu_{\text{eff}}^2}{3k_B}}
$


### Problema 2

En un material con anisotropía magnetocristalina cúbica, mostrar que para los coeficientes de orden dos vale $ b*{ii} = b*{jj} $, $ b\_{ij} = 0 $ si $ i \neq j $.  
La expresión para la energía es:

$
E_{MC} = E_0 + \sum_{i,j} b_{ij} \alpha_i \alpha_j + \sum_{i,j,k,l} b_{ijkl} \alpha_i \alpha_j \alpha_k \alpha_l + \mathcal{O}(6)
$


#### Respuesta

Tenemos que al ser un cristal cúbico, éste será invariante ante rotaciones de $90°$ _alrededor de los ejes_. Entonces, si rotamos, por ejemplo, alrededor del eje z:

$\alpha_1 \rightarrow \alpha_2, \hspace{1cm} \alpha_2 \rightarrow -\alpha_1 \hspace{1cm} \alpha_3 \rightarrow \alpha_3$

$
    E_{MC}(\alpha_1,\alpha_2,\alpha_3)=E_{MC}(\alpha_2,-\alpha_1,\alpha_3)
$
Tenemos que:

$
    E_{MC}(\alpha_1,\alpha_2,\alpha_3)= E_0 + b_{11}\alpha_1^2 + 2 b_{12}\alpha_1 \alpha_2 + ...+b_{22}\alpha_2^2+b_{33} \alpha_3^2 
$

$
    E_{MC}(\alpha_2,-\alpha_1,\alpha_3) = E_0 + b_{11}\alpha_1^2 - 2 b_{12}\alpha_1 \alpha_2 - ...+b_{22}\alpha_2^2+b_{33} \alpha_3^2 -...
$

$
    \longrightarrow 2b_{ij}\alpha_i \alpha_j =- 2b_{ij}\alpha_i \alpha_j \longrightarrow \boxed{b_{ij}=0}
$

$
    \boxed{b_{ii}=b_{jj}}
$


### Problema 3

En un cristal con simetría cúbica, la anisotropía magnetocristalina puede escribirse como:

$
E_{MC} = K_0 + K_1(\alpha_1^2\alpha_2^2 + \alpha_2^2\alpha_3^2 + \alpha_3^2\alpha_1^2) + K_2\alpha_1^2\alpha_2^2\alpha_3^2 + \dots
$
Determinar cuáles direcciones cristalinas corresponden a ejes fáciles, medios y difíciles para $ K_1 > 0 $, $ K_1 < 0 $ y distintos valores de $ K_2 $.


#### Respuesta

Tenemos que para las direcciones cristalográficas los cosenos directores valen:

$$
        \begin{split}
            [1,0,0]& \rightarrow (\alpha_1, \alpha_2, \alpha_3) = (1,0,0)\\
        [1,1,0]& \rightarrow (\alpha_1, \alpha_2, \alpha_3) = (\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0)\\
        [1,1,1] &\rightarrow (\alpha_1, \alpha_2, \alpha_3) = (\frac{1}{\sqrt{3}}, \frac{1}{\sqrt{3}}, \frac{1}{\sqrt{3}})
        \end{split}
$$

Entonces la energía magnetocristalina para cada una de las direcciones nos queda como:

$$
    E_{MC}^{[100]} = K_0 + K_1 (1 \cdot 0 + 0 \cdot 0 + 0 \cdot 1) + K_2 \cdot 1 \cdot 0 \cdot 0 = K_0
$$

$$
    E_{EM}^{[110]}= K_0 + K_1\left(\frac{1}{2} \cdot \frac{1}{2} + \frac{1}{2}\cdot 0 + 0 \cdot \frac{1}{2}\right) + K_2 \frac{1}{2} \cdot \frac{1}{2} \cdot 0 = K_0 + \frac{1}{4}K_1
$$

$$
    E_{MC}^{[111]}= K_0 + K_1 \left(\frac{1}{3}\cdot \frac{1}{3}+ \frac{1}{3}\cdot \frac{1}{3} +\frac{1}{3}\cdot \frac{1}{3} \right) +K_2 \frac{1}{3}\cdot \frac{1}{3} \cdot\frac{1}{3}= K_0 + \frac{1}{3}K_1 + \frac{1}{27}K_2
$$

$$
    \boxed{\begin{split}
        E_{MC}^{[100]}&=K_0\\
        E_{MC}^{[110]}&=K_0 + \frac{1}{4}K_1\\
        E_{MC}^{[111]}&=K_0 + \frac{1}{3}K_1 + \frac{1}{27}K_2
    \end{split}}
$$

Como podemos ver, si $K_1>0$ y $K_2 > 0$, tendremos siempre:

$$
    E_{MC}^{[100]}< E_{MC}^{[110]} < E_{MC}^{[111]}
$$

$\longrightarrow [1,0,0]$ plano fácil.

$\longrightarrow [1,1,0]$ plano medio.

$\longrightarrow [1,1,1]$ plano difícil.

Luego, vemos para qué valores de $K_2$ $E_{MC}^{[111]}<E_{MC}^{[110]}$:

$$
    \begin{split}
        K_0 + \frac{1}{3}K_1 - \frac{1}{27}|K_2| &< K_0 + \frac{1}{4}K_1\\
        \frac{1}{3}K_1 - \frac{1}{4} K_1 &< \frac{1}{27}|K_2|\\
        \frac{1}{12}K_1&<\frac{1}{27}|K_2|\\
        \longrightarrow |K_2| &< \frac{9}{4} K_1 \longrightarrow \boxed{K_2 > -\frac{9}{4}K_1}
    \end{split}
$$

Con lo que vemos que si $K_1 > 0$ y $K_2 < -\frac{9}{4} K_1$:

$\longrightarrow [1,0,0]$ plano fácil.

$\longrightarrow [1,1,1]$ plano medio.

$\longrightarrow [1,1,0]$ plano difícil.

Ahora analizamos en qué caso $E_{MC}^{[111]}<E_{MC}^{[100]}$, cuando $K_1>0$ y $K_2<0$:

$$
    \begin{split}
        K_0 + \frac{1}{3}K_1 - \frac{1}{27}|K_2| & < K_0\\
        \frac{1}{3}K_1 &< \frac{1}{27} |K_2|\\
        \longrightarrow |K_2| &> 9K_1 \longrightarrow\boxed{K_2 < -9K_1}
    \end{split}
$$

Con lo que vemos que si $K_1 >0$ y $K_2 < -9K_1$:

$\longrightarrow [1,1,1]$ plano fácil.

$\longrightarrow [1,1,0]$ plano medio.

$\longrightarrow [1,0,0]$ plano difícil.


```python
import numpy as np
import matplotlib.pyplot as plt

def alpha1(theta, phi):
    return np.cos(phi) * np.sin(theta)

def alpha2(theta, phi):
    return np.sin(phi) * np.sin(theta)

def alpha3(theta):
    return np.cos(theta)

def E_MC(theta, phi, K1=1, K2=0):
    a1 = alpha1(theta, phi)
    a2 = alpha2(theta, phi)
    a3 = alpha3(theta)

    term1 = a1**2 * a2**2
    term2 = a2**2 * a3**2
    term3 = a3**2 * a1**2
    term4 = a1**2 * a2**2 * a3**2

    return K1 * (term1 + term2 + term3) + K2 * term4

# --- Parámetros ---
K1 = 1
K2 = -15

# --- Malla ---
theta = np.linspace(0, np.pi, 200)
phi = np.linspace(0, 2 * np.pi, 200)
Theta, Phi = np.meshgrid(theta, phi)

# --- Energía ---
E = E_MC(Theta, Phi, K1=K1, K2=K2)

# --- Gráfico ---
plt.figure(figsize=(8, 6))
contour = plt.contourf(Phi, Theta, E, 100, cmap='viridis')

# Direcciones cristalinas
theta_100 = np.pi / 2
phi_100 = 0

theta_110 = np.pi / 2
phi_110 = np.pi / 4

theta_111 = np.arccos(1 / np.sqrt(3))
phi_111 = np.pi / 4

# Marcar puntos
plt.plot(phi_100, theta_100, 'ro', label='[100]')
plt.plot(phi_110, theta_110, 'go', label='[110]')
plt.plot(phi_111, theta_111, 'bo', label='[111]')

plt.legend(loc='upper right')

plt.xlabel(r'$\phi$ (rad)')
plt.ylabel(r'$\theta$ (rad)')


# Título indicando los valores de K1 y K2
plt.title(rf'Energía de anisotropía magnetocristalina ($K_1={K1}$, $K_2={K2}$)')

# Barra de color
plt.colorbar(contour, label='Energía')

plt.tight_layout()
plt.show()
```

### Problema 4

Evaluar si en un sistema con anisotropía uniaxial puede darse la situación de anisotropía "cónica", es decir que exista un ángulo de equilibrio distinto de $ \theta = 0 $ o $ \theta = \frac{\pi}{2} $.  
Para ello considerar hasta el término de orden 2 en la energía:

$
E = K_1 \sin^2 \theta + K_2 \sin^4 \theta
$

derivarla respecto a la variable angular y analizar en qué condiciones puede haber una solución que minimice la energía en un ángulo intermedio.


#### Respuesta


```python
import sympy as sp

# Variables simbólicas
theta = sp.symbols('theta', real=True)
K1, K2 = sp.symbols('K1 K2', real=True)

term1 = sp.sin(theta)

E = K1 * term1**2 + K2 * term1**4  

dE_dtheta = sp.diff(E, theta)

print("\nDerivada de E:")
sp.pprint(dE_dtheta)


dE2_dtheta2 = sp.diff(dE_dtheta, theta)

print("\nDerivada segunda de E:")
sp.pprint(sp.simplify(dE2_dtheta2))
```

$$
\begin{split}
\frac{dE_{MC}}{d\theta} & = 2K_1 \sin\theta \cos\theta + 4K_2 \sin^3\theta \cos\theta = 0 \\
& = 2 \cos\theta (K_1 \sin\theta + 2 K_2 \sin^3\theta) = 0
\end{split}
$$

$$
\Rightarrow K_1 \sin\theta + 2K_2 \sin^3\theta = 0
$$

$$
\sin\theta (K_1 + 2K_2 \sin^2\theta) = 0
\Rightarrow K_1 + 2 K_2 \sin^2\theta = 0
$$

$$
2 K_2 \sin^2\theta = -K_1
\Rightarrow \sin^2\theta = \frac{-K_1}{2K_2} \leq 1
\Rightarrow \boxed{|K_2|\geq \frac{|K_1|}{2}}
$$

Y tenemos que si $ K_1 > 0 \Rightarrow K_2 < 0 $ y si $ K_1 < 0 \Rightarrow K_2 > 0 $

---

Luego, analizamos la segunda derivada para verificar cuándo estos puntos son mínimos:

$$
\frac{d^2E_{MC}}{d\theta^2} = -4K_1 \sin^2\theta + 2K_1 - 16K_2 \sin^4\theta + 12K_2 \sin^2\theta
$$

Usamos que $ \sin^2\theta = \frac{-K_1}{2K_2} $:

$$
\begin{split}
\frac{d^2E_{MC}}{d\theta^2} & = -4K_1 \left( \frac{-K_1}{2K_2} \right)^2 + 2K_1
- 16K_2 \left( \frac{-K_1}{2K_2} \right)^4 + 12K_2 \left( \frac{-K_1}{2K_2} \right)^2 \\
& = \frac{4K_1^2}{2K_2} + 2K_1 - 4 \left( \frac{K_1^2}{K_2} \right) - 6K_1 \\
& = -\frac{2K_1^2}{K_2} - 4K_1 > 0
\Rightarrow \boxed{-\frac{2K_1^2}{K_2} > 4K_1}
\end{split}
$$

Luego, si suponemos $ K_1 > 0 \Rightarrow K_2 < 0 $ y $ |K_2| < \frac{|K_1|}{2} $:

$$
-\frac{2K_1^2}{K_2} = \frac{2K_1^2}{|K_2|} < \frac{2K_1^2}{\frac{K_1}{2}} = 4K_1
\Rightarrow -\frac{2K_1^2}{K_2} - 4K_1 < 0
$$

Lo que contradice la condición de mínimo. Por lo tanto:

$$
\boxed{\text{Para que se dé la anisotropía cónica: } K_1 < 0 \quad \text{y} \quad K_2 > \frac{-K_1}{2}}
$$


```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# Variables simbólicas
theta = sp.symbols('theta', real=True)
K1_val = -4
K2_val = 3

# Definir la energía simbólicamente
E_sym = K1_val * sp.sin(theta)**2 + K2_val * sp.sin(theta)**4

# Derivar E respecto a theta
dE_dtheta = sp.diff(E_sym, theta)

# Resolver dE/dtheta = 0
soluciones = sp.solve(dE_dtheta, theta, domain=sp.Interval(0, sp.pi))
print("\nSoluciones de dE/dtheta = 0 en [0, pi]:")
theta_criticos = []

for sol in soluciones:
    try:
        valor = float(sp.N(sol))
        if 0 <= valor <= np.pi:
            theta_criticos.append(valor)
            print(f"θ ≈ {valor:.4f} rad ({np.degrees(valor):.2f}°)")
    except TypeError:
        pass

# Crear función numérica para E y su derivada
E_func = sp.lambdify(theta, E_sym, modules=["numpy"])
dE_func = sp.lambdify(theta, dE_dtheta, modules=["numpy"])

# Crear un rango de theta
theta_vals = np.linspace(0, np.pi, 500)
E_vals = E_func(theta_vals)
dE_vals = dE_func(theta_vals)

# --- Graficar E(theta) ---
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(theta_vals, E_vals, label="Energía")
for tcrit in theta_criticos:
    plt.axvline(tcrit, color='red', linestyle='--', label=fr'$\theta_c={tcrit:.2f}$ rad')
plt.xlabel(r"$\theta$ (rad)")
plt.ylabel("Energía")
plt.title(fr"Energía $E(\theta)$ ($K_1={K1_val}$, $K_2={K2_val}$)")
plt.grid(True)
plt.legend()
plt.xticks([0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi],
           [r"$0$", r"$\frac{\pi}{4}$", r"$\frac{\pi}{2}$", r"$\frac{3\pi}{4}$", r"$\pi$"])

plt.tight_layout()
plt.show()
```

### Problema 5

¿Qué relación se debe cumplir entre los ejes $c$ y $a$ para que la anisotropía de forma de un elipsoide de revolución sea igual a la magnetocristalina?

**a)** ¿Se obtiene el mismo resultado para un elipsoide alargado que para uno achatado? Particularizar para el caso del cobalto ($ M_S \approx 1400\ \text{emu/cm}^3 $, $ K_1 \approx 5 \times 10^6\ \text{erg/cm}^3 $).


#### Respuesta

$
E_{\text{MS}} = \frac{1}{2} \vec{M} \tilde{N} \vec{M}
$

$
E_{\text{MS}} = \frac{1}{2}
\begin{pmatrix}
N_a & 0 & 0 \\
0 & N_b & 0 \\
0 & 0 & N_c
\end{pmatrix}
M^2 (\sin^2\varphi \sin^2\theta, \sin^2\theta \cos^2\varphi, \cos^2\theta)
$

$
E_{\text{MS}} = \frac{1}{2} M^2 (N_a \sin^2\varphi \sin^2\theta + N_b \sin^2\theta \cos^2\varphi + N_c \cos^2\theta)
$

Como $ a = b \Rightarrow N_a = N_b \Rightarrow 2N_a + N_c = 4\pi $:

$
E_{\text{MS}} = \frac{1}{2} M^2 \left( N_a \sin^2\varphi \sin^2\theta + N_a \sin^2\theta \cos^2\varphi + N_c \cos^2\theta \right)
$

$
= \frac{1}{2} M^2 \left[ N_a \sin^2\theta (\sin^2\varphi + \cos^2\varphi) + N_c \cos^2\theta \right]
$

$
= \frac{1}{2} M^2 \left( N_a \sin^2\theta + N_c (1 - \sin^2\theta) \right)
$

$
= \frac{1}{2} M^2 \left[ \sin^2\theta (N_a - N_c) + N_c \right]
$

$
= \frac{1}{2} M^2 (N_a - N_c) \sin^2\theta + \frac{1}{2} M^2 N_c
$

Prolato:

$
= \frac{1}{2} M^2 \left( \frac{4\pi - N_c - N_c}{2} \right) \sin^2\theta
= \frac{1}{2} M^2 \left( \frac{2\pi - 3N_c}{2} \right) \sin^2\theta
$

$
= M^2 \left( \pi - \frac{3}{4}N_c \right) \sin^2\theta
$

$
\Rightarrow \boxed{K_{\text{eff}} = M^2 \left( \pi - \frac{3}{4}N_c \right)}
$

$
\Rightarrow M^2 \left( \pi - \frac{3}{4}N_c \right) = K_1
$

$
\Rightarrow -\frac{3}{4}N_c M^2 = K_1 - M^2 \pi
\Rightarrow \boxed{N_c = -\frac{4K_1}{3M^2} + \frac{4}{3}\pi}
$

Oblato:

$
E_{\text{MS}} = \frac{1}{2} M^2 (N_a - 4\pi + 2N_a) \sin^2\theta
= \frac{1}{2} M^2 (3N_a - 4\pi) \sin^2\theta
$

$
= M^2 \left( \frac{3N_a}{2} - 2\pi \right) \sin^2\theta
\Rightarrow \boxed{K_{\text{eff}} = M^2 \left( \frac{N_a}{2} - 2\pi \right)}
$

$
\Rightarrow M^2 \left( \frac{N_a}{2} - 2\pi \right) = K_1
\Rightarrow M^2 N_a = 2K_1 + 4\pi M^2
\Rightarrow \boxed{N_a = \frac{2K_1 + 4\pi M^2}{M^2}}
$


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

# r = c/a
r = np.linspace(1.01, 5, 200)    # r > 1


# Parámetros del problema
Ms = 1400  # emu/cm³
K1 = 5e6   # erg/cm³

# Anisotropía de forma - prolato

Nc = (4 * np.pi)/(r**2 - 1) * ( r / (r**2 - 1)**(1/2) * np.log( r + (r**2 -1)**(1/2)) - 1)


K_forma_prolato = Ms**2 * (np.pi - (3/4) * Nc)

# r interseccion
Nc_int = - 4/3 * K1 / Ms**2 + 4/3 *np.pi

def Nc_prolato(r):
    return (4 * np.pi)/(r**2 - 1) * ( r / (r**2 - 1)**(1/2) * np.log( r + (r**2 -1)**(1/2)) - 1) - Nc_int

r_int_pro = fsolve(Nc_prolato, x0=4)
print("r intersección prolato = ", r_int_pro)

# Anisotropía de forma - oblato

Na = (4 * np.pi * r**2)/(r**2 -1) * (1 - 1/((r**2 -1)**(1/2)) * np.arcsin(((r**2 - 1)**(1/2))/r))

K_forma_oblato = np.abs(Ms**2 * ((Na)/2 - 2*np.pi))

# r intersección

Na_int = ((2 * K1) / Ms**2) + 4 * np.pi

def Na_oblato(r):
    Na_val = (4 * np.pi * r**2)/(r**2 -1) * (1 - 1/np.sqrt(r**2 -1) * np.arcsin(np.sqrt(r**2 - 1)/r))
    K_forma = Ms**2 * (Na_val/2 - 2*np.pi)
    return np.abs(K_forma) - K1

r_int_ob = fsolve(Na_oblato, x0=2)
print("r intersección oblato =", r_int_ob)

# Anisotropía magnetocristalina constante
K_cristalina = K1

# Graficar
plt.figure(figsize=(9, 5))
plt.plot(r, K_forma_prolato, label="K forma (prolato)", color="orange")
plt.plot(r, K_forma_oblato, label="K forma (oblato)", color="blue")
plt.axhline(K_cristalina, color="red", linestyle="--", label="K magnetocristalina")

# Estética
plt.xlabel(r"$r$")
plt.ylabel(r"$K_{\mathrm{efectiva}}$ (erg/cm³)")
plt.title("Comparación entre K de forma y K magnetocristalina")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Problema 6

Si el factor $ c/a $ entre el eje largo y el corto es el mismo, ¿cuál tiene más anisotropía de forma, el elipsoide alargado o el achatado?


#### Respuesta

Para el elipsoide achatado tenemos:

$$
K_{\text{eff}} = M^2 \left( \frac{N_a}{2} - 2\pi \right)
$$

Si es muy achatado, entonces $N_a = N_b \approx 0$, $ N_c \approx 4\pi$

$$
\Rightarrow K_{\text{eff}} = M^2 \left(0 - 2\pi \right) = \boxed{- 2 \pi M^2}
$$

---

Para el elipsoide alargado tenemos:

$$
K_{\text{eff}} = M^2 \left( \pi - \frac{3}{4} N_c \right)
$$

Si es muy alargado $\Rightarrow N_a = N_b \approx 2\pi$, $N_c \approx 0$:

$$
K_{\text{eff}} = M^2 (\pi - 0) = \boxed{M^2 \pi}
$$

---

Con lo que podemos ver que el achatado posee mayor anisotropía de forma.


### Problema 7

¿En qué dirección estará el eje fácil si el ángulo entre la dirección del eje dado por la anisotropía de forma y el eje magnetocristalino uniaxial es distinto de $ \frac{\pi}{2} $? Particularizar para el caso $ K*{\text{forma}} = K*{\text{mc}} $.


Partimos de la forma de la energía:

$$
E = K_{\text{forma}} \sin^2\theta + K_{\text{mc}} \sin^2(\varphi - \theta)
$$

Derivamos respecto a $\theta$:

$$
\frac{dE}{d\theta} = 2K_{\text{forma}} \sin\theta \cos\theta - 2K_{\text{mc}} \sin(\varphi - \theta)\cos(\varphi - \theta)
$$

Usamos identidades trigonométricas:

$$
= K_{\text{forma}} \sin 2\theta - K_{\text{mc}} \sin 2(\varphi - \theta)
$$

Aplicamos la fórmula del ángulo doble:

$$
\sin 2(\varphi - \theta) = \sin 2\varphi \cos 2\theta - \cos 2\varphi \sin 2\theta
$$

Entonces:

$$
\frac{dE}{d\theta} = K_{\text{forma}} \sin 2\theta - K_{\text{mc}} (\sin 2\varphi \cos 2\theta - \cos 2\varphi \sin 2\theta)
$$

$$
= \sin 2\theta (K_{\text{forma}} + K_{\text{mc}} \cos 2\varphi) - K_{\text{mc}} \sin 2\varphi \cos 2\theta = 0
$$

Agrupamos:

$$
\sin 2\theta (K_{\text{forma}} + K_{\text{mc}} \cos 2\varphi) = K_{\text{mc}} \sin 2\varphi \cos 2\theta
$$

Pasamos dividiendo el $\cos 2\theta$ (si $\cos 2\theta \ne 0$):

$$
\tan 2\theta = \frac{\sin 2\theta}{\cos 2\theta} = \frac{K_{\text{mc}} \sin 2\varphi}{K_{\text{mc}} \cos 2\varphi + K_{\text{forma}}}
$$

O también:

$$
\tan 2\theta = \frac{\sin 2\varphi}{\frac{K_{\text{forma}}}{K_{\text{mc}}} + \cos 2\varphi}
$$

Caso particular: si $K_{\text{forma}} = K_{\text{mc}}$, entonces:

$$
\tan 2\theta = \tan \varphi
$$

### Problema 9

En clase vimos que una de las formas de determinar la anisotropía efectiva en una película delgada (dentro del modelo de Stoner-Wohlfarth) es a partir de la diferencia de áreas entre los lazos $ M $ vs. $ H $ medidos en las direcciones paralela y perpendicular al plano de la película, $ A*{\parallel} - A*{\perp} = K*{\text{eff}} $. Experimentalmente suele ser complicado alinear el film, especialmente en la configuración perpendicular, sin embargo el error que se comete no es (no debería ser...) muy significativo. Mostrar a partir de cálculos numéricos que para cometer un error del $1\%$ en la determinación de $ K*{\text{eff}} $ es necesario que la muestra esté desalineada por lo menos 5°.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar
from scipy.integrate import quad
import pandas as pd

# --- Parámetros físicos ---
K = 5e6   # erg/cm³
M = 1400 # emu/cm³

# --- Ángulos alpha en grados (para el gráfico) y radianes ---
alpha_grados = [0, 80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90]
alphas = [np.deg2rad(a) for a in alpha_grados]

# --- Rango de campo H en Oe ---
H_vals = np.linspace(0, 2*K/M, 500)

# --- Energía total ---
def energia(theta, H, alpha):
    return -H * M * np.cos(theta - alpha) + K * np.sin(theta)**2

# --- Minimiza la energía y encuentra θ_eq ---
def theta_equilibrio(H, alpha):
    res = minimize_scalar(energia, bounds=(0, np.pi), args=(H, alpha), method='bounded')
    return res.x


# --- Cálculo de curvas M_H y áreas ---
areas = []
MH_dict = {}

for alpha, a_deg in zip(alphas, alpha_grados):
    MHs = []
    for H in H_vals:
        theta = theta_equilibrio(H, alpha)
        Mh = M * np.cos(theta - alpha)
        MHs.append(Mh)
    MH_dict[a_deg] = MHs
    area = quad(lambda H: np.interp(H, H_vals, MHs), H_vals[0], H_vals[-1],
                limit=3000, epsabs=1e-13, epsrel=1e-13)[0]
    areas.append(area)

# --- Cálculo del delta de área y errores relativos ---
area_paralela = areas[0]        # α = 0°, eje fácil 
area_perpendicular = areas[-1]  # α = 90°, eje difícil 
delta_max = area_paralela - area_perpendicular



deltas = [abs(delta_max - (area_paralela - a )) for a in areas]
error_relativo = [delta / area_paralela * 100 for delta in deltas]

# --- Mostrar tabla ---
df = pd.DataFrame({
    "Alpha (°)": alpha_grados,
    "Área": areas,
    "Error (mod)": deltas,
    "Error porcentual": error_relativo
})

print(df)

# --- Gráfico de curvas M_H ---
plt.figure(figsize=(8, 6))
for a_deg in alpha_grados:
    plt.plot(H_vals, MH_dict[a_deg], label=fr"$\alpha = {a_deg}^\circ$")
plt.xlabel("Campo aplicado $H$ (Oe)")
plt.ylabel("Magnetización proyectada $M_H$")
plt.title("Curvas $M_H$ vs $H$ para distintos ángulos $\\alpha$")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Problema 10

Se tiene un sistema uniaxial en el que se aplica un campo $ H $ a un ángulo $\alpha$ . El campo es lo suficientemente grande como para alcanzar la saturación. Si se reduce el valor de $ H $ (haciéndose eventualmente negativo) en general se observa que para algún campo crítico la magnetización pega un salto discontinuo y tiende a alinearse con $ H $. Para $ \alpha = 0 $ ó $ \pi/2 $ este campo vale $ 2K/M_s $.

**a)** ¿Cuánto vale el campo crítico para ángulos intermedios?

**b)** ¿Para qué $ \alpha $ se produce el campo crítico mínimo?


#### Respuesta

**a)**

La energía del sistema uniaxial con un campo aplicado en una dirección arbitraria $\alpha$ es:

$$E(\theta) = -HM \cos(\theta - \alpha) + K \sin^2\theta$$

Queremos determinar el campo crítico $H_c$ en función del ángulo $\alpha$. El campo crítico es aquel en el que la magnetización sufre un salto discontinuo (conmutación), es decir, el punto en el que desaparece el mínimo local de la energía.

Esto ocurre cuando la energía tiene un punto de inflexión degenerado, es decir, cuando:

$$\frac{dE}{d\theta} = 0 \quad \text{y} \quad \frac{d^2E}{d\theta^2} = 0$$

Calculamos la primera derivada:

$$\frac{dE}{d\theta} = HM \sin(\theta - \alpha) + 2K \sin\theta \cos\theta$$

y la segunda derivada:

$$\frac{d^2E}{d\theta^2} = HM \cos(\theta - \alpha) + 2K (\cos^2\theta - \sin^2\theta)$$

Condiciones para el campo crítico:

Buscamos los valores de $\theta$ y $H$ tales que:

$$\frac{dE}{d\theta} = 0 \quad \text{y} \quad \frac{d^2E}{d\theta^2} = 0$$

Llamamos a estas ecuaciones (1) y (2):

$$
\begin{align}
&\text{(1)} \quad HM \sin(\theta - \alpha) + 2K \sin\theta \cos\theta = 0 \\
&\text{(2)} \quad HM \cos(\theta - \alpha) + 2K (\cos^2\theta - \sin^2\theta) = 0
\end{align}
$$

Definimos $x = \cos\theta$ y $y = \sin\theta$, y usamos que $x^2 + y^2 = 1$. Usamos identidades trigonométricas:

$$\sin(\theta - \alpha) = \sin\theta \cos\alpha - \cos\theta \sin\alpha = y \cos\alpha - x \sin\alpha$$

$$\cos(\theta - \alpha) = \cos\theta \cos\alpha + \sin\theta \sin\alpha = x \cos\alpha + y \sin\alpha$$

Reemplazamos en (1):

$$HM (y \cos\alpha - x \sin\alpha) + 2K x y = 0 \tag{3}$$

Y en (2):

$$HM (x \cos\alpha + y \sin\alpha) + 2K (x^2 - y^2) = 0 \tag{4}$$

Resolviendo el sistema:

Multiplicamos (3) por $\cos\alpha$ y (4) por $\sin\alpha$, y sumamos:

$$
\begin{align*}
&HM(y \cos^2\alpha - x \sin\alpha \cos\alpha) + 2K x y \cos\alpha \\
+& HM(x \cos\alpha \sin\alpha + y \sin^2\alpha) + 2K(x^2 - y^2)\sin\alpha = 0
\end{align*}
$$

Agrupamos términos en $x$ y $y$:

$$
HM y (\cos^2\alpha + \sin^2\alpha) + HM x ( \cos\alpha \sin\alpha - \sin\alpha \cos\alpha)
+ 2K x y \cos\alpha + 2K(x^2 - y^2)\sin\alpha = 0
$$

Notamos que $\cos^2\alpha + \sin^2\alpha = 1$ y el segundo término se anula, por lo que queda:

$$HM y + 2K x y \cos\alpha + 2K(x^2 - y^2)\sin\alpha = 0$$

Factorizamos $y$:

$$y \left[HM + 2K x \cos\alpha + 2K \left( \frac{x^2 - (1 - x^2)}{y} \right) \sin\alpha \right] = 0$$

Finalmente, se obtiene una expresión analítica que nos lleva a:

$$\frac{H_c M}{2K} = \left( \cos^{2/3}\alpha + \sin^{2/3}\alpha \right)^{-3/2}$$

Este resultado permite evaluar el campo crítico reducido para cualquier ángulo $\alpha$ entre $0^\circ$ y $90^\circ$. Se puede graficar esta expresión para visualizar que el mínimo ocurre en $\alpha = 45^\circ$:

$$\boxed{ \min\left(\frac{H_c M}{2K}\right) = \frac{1}{\sqrt{2}} \approx 0.707 \quad \text{para } \alpha = 45^\circ }$$


### Problema 11

En un sistema con anisotropía uniaxial la energía magnética puede desarrollarse a cuarto orden:

$$E = -HM \cos(\theta - \alpha) + K_1 \sin^2\theta + K_2 \sin^4\theta$$

**a)** En el caso en que $K_1 > 0 $, dibuje la curva de magnetización en función del campo magnético cuando $H$ se aplica perpendicular al eje fácil en los casos en que el parámetro $K_2$ es menor, igual o mayor que 0.

**b)** La dependencia funcional entre $M$ y $H$ puede linealizarse usando el formalismo de Sucksmith-Thompson en el que se grafica $H/M_{\perp}$ vs. $M_{\perp}^2$. A partir de la ecuación anterior diga cómo determinar $K_1$ y $K_2$.


#### Respuestas

**a)**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import newton

# --- Parámetros físicos ---
K1 = 5e6       # Término cuadrático de anisotropía (erg/cm³)
K2 = float(input("Ingrese K2 (ej. -2.5e6, 0, 1e6): "))  # Término cuártico de anisotropía
M = 1400       # Magnetización de saturación (emu/cm³)
Hk_ref = 2*K1/M  # Campo de anisotropía de referencia

# --- Configuración del barrido ---
N = 1000
H_max = 1.5 * max(2*K1 + 4*abs(K2), 2*K1)/M  # Rango dinámico seguro
H_vals_down = np.linspace(H_max, -H_max, N)
H_vals_up = np.linspace(-H_max, H_max, N)

# --- Energía y derivadas CORRECTAS ---
def energy(theta, H):
    """Energía total del sistema"""
    return -H*M*np.sin(theta) + K1*np.sin(theta)**2 + K2*np.sin(theta)**4

def dE_dtheta(theta, H):
    """Primera derivada de la energía respecto a theta"""
    return (-H*M*np.cos(theta) + 
            2*K1*np.sin(theta)*np.cos(theta) + 
            4*K2*np.sin(theta)**3*np.cos(theta))

def d2E_dtheta2(theta, H):
    """Segunda derivada (para verificar mínimos)"""
    return (H*M*np.sin(theta) + 
            2*K1*np.cos(2*theta) + 
            4*K2*(3*np.sin(theta)**2*np.cos(theta)**2 - np.sin(theta)**4))

# --- Función de solución física MEJORADA ---
def find_physical_solution(H, previous_theta):
    """Encuentra la solución física para un campo H dado"""
    # Determinamos los mínimos esperados según K2
    if K2 >= 0:
        # Para K2 positivo, mínimos cerca de 0 y π
        base_angles = [0, np.pi]
    else:
        # Para K2 negativo, mínimos cerca de ±π/2
        base_angles = [np.pi/2, -np.pi/2]
    
    # Generamos puntos iniciales inteligentes
    initial_guesses = (
        [previous_theta] +  # Mantener continuidad
        base_angles +       # Mínimos esperados
        [a + 0.1 for a in base_angles] +  # Variaciones positivas
        [a - 0.1 for a in base_angles]    # Variaciones negativas
    )
    
    solutions = []
    for guess in initial_guesses:
        try:
            theta = newton(dE_dtheta, x0=guess, args=(H,), tol=1e-8, maxiter=100)
            if d2E_dtheta2(theta, H) > 0:  # Verificar que es mínimo
                solutions.append(theta)
        except RuntimeError:
            continue
    
    # Seleccionamos la solución más cercana a la anterior
    return min(solutions, key=lambda x: abs(x - previous_theta)) if solutions else previous_theta

# --- Condiciones iniciales CORRECTAS ---
if K2 >= 0:
    theta_init_down = 0.01          # Cerca de θ=0 para barrido descendente
    theta_init_up = np.pi - 0.01    # Cerca de θ=π para barrido ascendente
else:
    theta_init_down = np.pi/2 + 0.01  # Cerca de θ=π/2 para K2 negativo
    theta_init_up = -np.pi/2 - 0.01   # Cerca de θ=-π/2 para K2 negativo

# --- Resolución de la histéresis ---
thetas_down = []
theta = theta_init_down
for H in H_vals_down:
    theta = find_physical_solution(H, theta)
    thetas_down.append(theta)

thetas_up = []
theta = theta_init_up
for H in H_vals_up:
    theta = find_physical_solution(H, theta)
    thetas_up.append(theta)

# --- Proyección de la magnetización ---
MH_down = M * np.sin(thetas_down)  # Proyección M·H/|H|
MH_up = M * np.sin(thetas_up)

# --- Visualización MEJORADA ---
plt.figure(figsize=(12, 7))
plt.plot(H_vals_down/Hk_ref, MH_down/M, 'b-', lw=2, label='Barrido descendente')
plt.plot(H_vals_up/Hk_ref, MH_up/M, 'r-', lw=2, label='Barrido ascendente')

plt.title(f'Modelo Stoner-Wohlfarth extendido\n$K_1$ = {K1/1e6}×10$^6$ erg/cm³, $K_2$ = {K2/1e6}×10$^6$ erg/cm³', pad=20)
plt.xlabel('Campo aplicado ($H/H_{k0}$)', fontsize=12)
plt.ylabel('Magnetización ($M/M_s$)', fontsize=12)
plt.grid(True, alpha=0.3)

# Líneas de referencia
plt.axhline(0, color='k', ls='--', lw=0.5)
plt.axvline(0, color='k', ls='--', lw=0.5)
plt.axhline(1, color='gray', ls=':', lw=0.5, alpha=0.5)
plt.axhline(-1, color='gray', ls=':', lw=0.5, alpha=0.5)

# Anotación informativa
plt.annotate(f'Configuración:\nα = 90°\n$H_{{k0}}$ = {Hk_ref:.1f} Oe',
             xy=(0.98, 0.05), xycoords='axes fraction', ha='right')

plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

#### Método de Sucksmith-Thompson para determinar $K_1$​ y $K_2$

1. Base Teórica

La energía del sistema para un campo aplicado perpendicular al eje fácil $(\alpha = pi/2)$ es:

$$E=−HM\sin\theta+K_1\sin^⁡2\theta+K_2\sin^⁡4\theta$$

En equilibrio $(dE/d\theta=0)$:

$$HM\cos⁡\theta=2K_1\sin⁡\theta\cos⁡\theta+4K_2\sin⁡^3\theta\cos⁡\theta$$

### Problema 14

En un cristal cúbico la magnetostricción de saturación puede escribirse como:

$$\epsilon_s = \frac{3}{2} \lambda_{100}\left(\alpha_1^2\beta_1^2 + \alpha_2^2\beta_2^2 + \alpha_3^2\beta_3^2 - \frac{1}{3}\right) + 3\lambda_{111}(\alpha_1\beta_1\alpha_2\beta_2 + \alpha_2\beta_2\alpha_3\beta_3 + \alpha_3\beta_3\alpha_1\beta_1)$$

donde $\lambda_{100}$ y $\lambda_{111}$ son los valores de saturación de la magnetostricción en los ejes cristalinos correspondientes, y $\alpha_i$ y $\beta_i$ son los cosenos directores de la magnetización $M$ y de la dirección en la que se mide la deformación $\epsilon$ . Obtenga una expresión para el coeficiente $\lambda_{110}$ en función de $\lambda_{100}$ y $\lambda_{111}$ en el caso en que la deformación se mide en la misma dirección que $M$.


#### Respuesta

La magnetostricción de saturación en un cristal cúbico viene dada por:

$$\epsilon_s = \frac{3}{2} \lambda_{100}\left(\sum_{i=1}^3 \alpha_i^2\beta_i^2 - \frac{1}{3}\right) + 3\lambda_{111}\left(\sum_{i \neq j} \alpha_i\beta_i\alpha_j\beta_j\right)$$

donde:

- $\lambda_{100}$, $\lambda_{111}$: Coeficientes de magnetostricción en las direcciones $[100]$ y $[111]$
- $\alpha_i$: Cosenos directores de la magnetización $\mathbf{M}$
- $\beta_i$: Cosenos directores de la dirección de medida $\epsilon$


- Caso particular: $\mathbf{M} \parallel [110]$ y medida en $[110]$
Para la dirección $[110]$:

$$\alpha_1 = \beta_1 = \frac{1}{\sqrt{2}}, \quad \alpha_2 = \beta_2 = \frac{1}{\sqrt{2}}, \quad \alpha_3 = \beta_3 = 0$$


Sustitución en la expresión general

$$\begin{align}
\epsilon_s &= \frac{3}{2}\lambda_{100}\left[\left(\frac{1}{2}\right)\left(\frac{1}{2}\right) + \left(\frac{1}{2}\right)\left(\frac{1}{2}\right) + 0 - \frac{1}{3}\right] \\
&\quad + 3\lambda_{111}\left[\left(\frac{1}{2}\right)\left(\frac{1}{2}\right) + 0 + 0\right] \\
&= \frac{3}{2}\lambda_{100}\left(\frac{1}{2} - \frac{1}{3}\right) + \frac{3}{4}\lambda_{111} \\
&= \frac{3}{2}\lambda_{100}\left(\frac{1}{6}\right) + \frac{3}{4}\lambda_{111} \\
&= \frac{\lambda_{100}}{4} + \frac{3\lambda_{111}}{4}
\end{align}$$

Definición de $\lambda_{110}$:

Cuando $\mathbf{M}$ y la dirección de medida están ambos en $[110]$, la deformación es precisamente $\lambda_{110}$:

$$\epsilon_s \equiv \lambda_{110}$$

$$\boxed{\lambda_{110} = \frac{1}{4}\lambda_{100} + \frac{3}{4}\lambda_{111}}$$


Interpretación física:


-$\lambda_{110}$ es un promedio ponderado de $\lambda_{100}$ y $\lambda_{111}$
- La dirección $[110]$ combina contribuciones de ambas constantes debido a su simetría intermedia
- Si $\lambda_{100} = \lambda_{111}$, entonces $\lambda_{110} = \lambda_{100}$ (caso isotrópico)
